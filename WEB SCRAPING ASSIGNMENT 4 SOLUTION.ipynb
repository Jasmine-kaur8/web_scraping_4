# Q1. Scrape the details of most viewed videos on YouTube from Wikipedia

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(6)

#creating empty lists
name=[]
artist=[]
upload_date = []
views=[]


#scraping names of the video



try:

    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")

    for i in names:

        name.append(i.text)

except NoSuchElementException:#handling no such element exception

    name.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    name.append('No details available')
    
time.sleep(4)
    
    
#scraping names of the artist
    
try:

    artist_names = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")

    for i in artist_names:

        artist.append(i.text)

except NoSuchElementException:#handling no such element exception

    name.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    name.append('No details available')
    
time.sleep(4)
    
#scraping date of uploading the video
try:

    date = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")

    for i in date:

        upload_date.append(i.text)

except NoSuchElementException:#handling no such element exception

    upload_date.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    upload_date.append('No details available')
    
#scraping the data of number of views in billions


try:

    view = driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")

    for i in view:

        views.append(i.text)

except NoSuchElementException:#handling no such element exception

    views.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    views.append('No details available')

        

time.sleep(4)

 # creating the dataframe 
df=pd.DataFrame({"Rank": range(0,30), "Name of the video":name,"Name of the artist":artist,"Upload date":upload_date,
                 "Views(in billions)":views})
df
```

```python
driver.close()
```

# Q2 Scrape the details team India’s international fixtures from bcci.tv

```python
driver=webdriver.Chrome("chromedriver.exe") 

url = "https://www.bcci.tv/"
driver.get(url)


#going to international fixtures page
from selenium.common.exceptions import ElementNotInteractableException

fixtures=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']/div/ul/li/a")

try:

    fixtures.click()

except ElementNotInteractableException:

    driver.get(fixtures.get_attribute('href'))
    
#creating empty lists
Match_title = []
Series = []
Place = []
Date = []
Time = []
```

```python
#scraping match title


try:

    title = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")

    for i in title:

        Match_title.append(i.text)

except NoSuchElementException:#handling no such element exception

    Match_title.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    Match_title.append('No details available')
    


    
#scraping series
    
try:

    ser = driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")

    for i in ser:

        Series.append(i.text)

except NoSuchElementException:#handling no such element exception

    Series.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    Series.append('No details available')
    

    
#scraping place
try:

    pl = driver.find_elements_by_xpath("//p[@class='fixture__additional-info'][1]/span")

    for i in pl:

        Place.append(i.text)

except NoSuchElementException:#handling no such element exception

    Place.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    Place.append('No details available')
    
Place
    

#scraping date


try:

    date =driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']/div/span")
    month=driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']/div/div/span[1]")
    j=len(date)
    for i in range(j):
        try:
            Date.append(date[i].text+" "+month[i].text)
        except:
            Date.append("-") 
    

except NoSuchElementException:#handling no such element exception

    Date.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    Date.append('No details available')

                                
#scraping time

try:

    time = driver.find_elements_by_xpath("//span[@class='fixture__time']")

    for i in time:

        Time.append(i.text)

except NoSuchElementException:#handling no such element exception

    Time.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    Time.append('No details available')

        
 # creating the dataframe 
df=pd.DataFrame({"Match Title":Match_title, "Series":Series,"Place":Place, "Time":Time,"Date":Date})
df



```

```python
driver.close()    #closing the driver
```

# Q4. Scrape the details of State-wise GDP of India from statisticstime.com

```python
#connecting to the web driver and opening the url

driver=webdriver.Chrome("chromedriver.exe") 

url = "http://statisticstimes.com/"
driver.get(url)

#opening the economy tab

economy = driver.find_element_by_xpath("//*[@id='top']/div[2]/div[2]/button")
economy.click()


#clicking the india tab

india = driver.find_element_by_xpath("//*[@id='top']/div[2]/div[2]/div/a[3]")
india.click()
```

```python
#clicking the state wise GDP of India

GDP = driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[2]/ul/li[1]/a")
GDP.click()


#creating empty lists

Rank = []
State = []
GSDP_20 = []
GSDP_19 = []
Share_19 = []
GDP_Billion = []
```

```python
#extracting names

n = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[1]")

for i in n:
    try:
        Rank.append(i.text)
    except:
        Rank.append("-")

        
#extracting state list

s = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[2]")
for i in s:
    try:
        State.append(i.text)
    except:
        State.append("-")

        
#extracting GSDP(19-20)

gdp = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[3]")
for i in gdp:
    try:
        GSDP_20.append(i.text)
    except:
        GSDP_20.append("-")

        
#extracting GSDP(18-19)

gdp_ = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]")
for i in gdp_:
    try:
        GSDP_19.append(i.text)
    except:
        GSDP_19.append("-")
        

#extracting share(18-19)

share = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[5]")
for i in share:
    try:
        Share_19.append(i.text)
    except:
        Share_19.append("-")

        
#extracting GDP(in $ billions)

gdp_1 = driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]")
for i in gdp_1:
    try:
        GDP_Billion.append(i.text)
    except:
        GDP_Billion.append("-")

        
#creating the datadrame

State_GDP = pd.DataFrame({})
State_GDP['RANK'] = Rank[0:297]
State_GDP['STATE'] = State[0:297]
State_GDP['GSDP at current price(19-20)'] = GSDP_20[0:297]
State_GDP['GSDP at current price(18-19)'] = GSDP_19[0:297]
State_GDP['Share(18-19)'] = Share_19[0:297]
State_GDP['GDP($ billion)'] = GDP_Billion[0:297]
State_GDP
```

```python
driver.close()
```

# Q5. Scrape the details of trending repositories on Github.com.


```python
#connecting to the web driver and opening the url

driver=webdriver.Chrome("chromedriver.exe") 

url = "https://github.com/"
driver.get(url)
```

```python
#clicking the explore tab

l = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")

l.click()

#clicking the trending tab

trending = driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a")
trending.click()
```

```python
#creating empty lists

Repository_title = []
Repository_description = []
Contributors_count = []
Language_used = []
url = []
```

```python
#extracting Repository title

rep = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in rep:
    try:
        Repository_title.append(i.text)
    except:
        Repository_title.append("-")
        
#Extracting Contributors_countlist
rep1 = driver.find_elements_by_xpath("//*[@id='js-pjax-container']/div[3]/div/div[2]/article/div[2]/a[2]")
for i in rep1:
    try:
        Contributors_count.append(i.text)
    except:
        Contributors_count.append("-")

        
# urls for Repository_desc & Language_usedinfo
link = driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a")
for i in link:
    url.append(i.get_attribute("href"))

    

#Extracting Repository description
for i in url:
    driver.get(i)
    try:
        rep1=driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p | //*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/div[1]')
        Repository_description.append(rep1.text)
    except NoSuchElementException as e:
            Repository_description.append("-")
            
            
#Extracting Language_used 
for i in url:
    driver.get(i)
    try:
        rep3=driver.find_element_by_xpath('//span[@class="color-text-primary text-bold mr-1"]')
        Language_used.append(rep3.text)
    except NoSuchElementException as e:
            Language_used.append("-")  

            

#creating the dataframe
Github=pd.DataFrame({})
Github['Repository Title']=Repository_title[0:25]
Github['Repository Description']=Repository_description[0:25]
Github['Contributors Count']=Contributors_count[0:25]
Github['Language Used']=Language_used[0:25]
Github
```

```python
#closing the driver
driver.close()
```

# Q10. . Details of Datasets from UCI machine learning repositories

```python
#connecting to the web driver and opening the url

driver=webdriver.Chrome("chromedriver.exe") 

url = "https://archive.ics.uci.edu/"
driver.get(url)
```

```python
#clicking dataset link

ds = driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
ds.click()
```

```python
#creating the empty list

Dataset_Name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
#Extracting Dataset_Name
d = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
for i in d:
    try:
        Dataset_Name.append(i.text)
    except:
        Dataset_Name.append("-")
        

#Extracting Data_type
dtype = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
for i in dtype[1:]:
    try:
        Data_type.append(i.text)
    except:
        Data_type.append("-")
        
#Extracting Task
task = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
for i in task[1:]:
    try:
        Task.append(i.text)
    except:
        Task.append("-")
        

#Extracting Attribute_type
att = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
for i in att[1:]:
    try:
        Attribute_type.append(i.text)
    except:
        Attribute_type.append("-")    
        
#Extracting Number of instances
instances = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
for i in instances[1:]:
    try:
        No_of_instances.append(i.text)
    except:
        No_of_instances.append("-")

#Extracting Number of attribute
attr = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
for i in attr[1:]:
    try:
        No_of_attribute.append(i.text)
    except:
        No_of_attribute.append("-")
        
#Extracting Year
year = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
for i in year[1:]:
    try:
        Year.append(i.text)
    except:
        Year.append("-")
        

#creating dataframe
dataset = pd.DataFrame({})
dataset['Dataset Name' ] = Dataset_Name[0:1763]
dataset['Data type'] = Data_type[0:1763]
dataset['Task'] = Task[0:1763]
dataset['Attribute type'] = Attribute_type[0:1763]
dataset['No of instances'] = No_of_instances[0:1763]
dataset['No of attribute'] = No_of_attribute[0:1763]
dataset['Year'] = Year[0:1763]
dataset
```
